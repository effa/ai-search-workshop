# Problém plánování

```python
%matplotlib inline
from search import parse_state, is_goal, actions, move, show_plan
```

Plánování si představíme na konkrténím problému *hledání cesty ve vesmíru*. Naším cílem bude sestavit plán pro raketku ve vesmíru,
podle kterého raketka doletí do cíle, aniž by po cestě narazila do asteroidu.

Vesmírné prostředí tvoří mřížka, ve které jsou některá políčka zablokovaná asteroidy. Raketka začíná na některém políčku na spodním řádku a musí doletět do horního řádku. Stav raketky je tedy určen její x-ovou a y-ovou pozicí, kterou budeme zkráceně uvádět v "šachové notaci", např. v následujícím stavu je raketka na pozici *c1*:

```python
state = parse_state(
'| | | | | |'
'| | | | | |'
'|A|A|A|A| |'   # A = asteroid
'| | | | | |'
'| | |S| | |')  # S = spaceship
state
```

## Testování cíle

*Cílový stav* je takový stav, ve kterém se raketka nachází na horním řádku. Cílovost stavu testuje funkce `is_goal(state)`.

```python
another_state = state.at('d5')
print('Je cilovy?', is_goal(another_state))
another_state
```

## Akce a následníci

Raketka letí stále vpřed a může přitom zatáčet doleva a doprava (dostoupné jsou tedy tři akce: 'l' jako 'left', 'f' jako 'forward', 'r' jako 'right'. Seznam akcí dostupných v daném stavu vrací funkce `actions(state)`:

```python
print('Dostupne akce:', actions(state))
state
```

Funkce `move(s, a)` vrací stav, který vznikne ze stavu *s* po provedení akce *a*:

```python
next_state = move(state, 'r')
next_state
```

Všimněte si, že raketka se při akci `r` posunula nejen o 1 pole doprava, ale současně i o 1 pole vpřed (tj. raketka letí stále vpřed).

V některých stavech nejsou dostupné všechny tři akce, ale jen některé:

```python
actions(next_state)
```
